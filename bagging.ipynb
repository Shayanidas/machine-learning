```python
x_train, y_train = train.drop(['label'], axis =1), train['label']
x_test, y_test = test.drop(['label'],axis =1), test['label']
```

```python
#check x_train dataset
x_train
```

```python
#check y_train dataset
y_train
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(x_train, y_train)
Linear_Regression = lreg.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
LogisticRegression = logreg.score(x_test, y_test)
```

```python
# apply k nearest neighborhood classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(x_train, y_train)
KNN_Classifier = knn.score(x_test, y_test)
```

```python
# apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
Decision_Tree_Classifier = dtc.score(x_test, y_test)
```

```python
# apply random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10)
rfc.fit(x_train, y_train)
Random_Forest_Classifier = rfc.score(x_test, y_test)
```

```python
# apply bagging with decision tree classifier
from sklearn.ensemble import BaggingClassifier
bg_dtc = BaggingClassifier(DecisionTreeClassifier(), max_samples = 0.7, n_estimators = 10)
bg_dtc.fit(x_train, y_train)
Bagging_DTC = bg_dtc.score(x_test, y_test)
```

```python
# apply bagging with decision tree classifier
bg_rfc = BaggingClassifier(RandomForestClassifier(n_estimators = 10), max_samples = 0.7, n_estimators = 10)
bg_rfc.fit(x_train, y_train)
Bagging_RFC = bg_rfc.score(x_test, y_test)
```

```python
# compare accuracy of models in different algorithm
import matplotlib.pyplot as plt
data = {'Logistic Regression': LogisticRegression, 'KNN': KNN_Classifier,'Decision Tree Classifier': Decision_Tree_Classifier, 'Random Forest Classifier': Random_Forest_Classifier, 'Bagging_DTC': Bagging_DTC, 'Bagging_RFC': Bagging_RFC}
algo = data.keys()
scores = data.values()
plt.bar(algo, scores, color ='maroon', width = 0.4)
plt.xticks(rotation=90)
plt.ylabel("Score")
plt.title("ML Algorithm Comparison")
plt.show()
```
